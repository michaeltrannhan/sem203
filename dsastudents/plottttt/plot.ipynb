```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

besttime = pd.read_csv('besttime.csv')
#print(besttime.iloc[0:3])

besttime = besttime.to_numpy()

plt.figure(figsize=(15,8))

plt.scatter(besttime[:, 0],  besttime[:,1], c='r', marker='d', alpha=0.9, label='InsertionSort')
plt.scatter(besttime[:, 0],  besttime[:,2], c='g', marker='s', alpha=0.9, label='ShellSort')
plt.scatter(besttime[:, 0],  besttime[:,3], c='navy', marker='.', alpha=0.9, label='SelectionSort')
plt.scatter(besttime[:, 0],  besttime[:,4], c='b', marker='o', alpha=0.9, label='BubbleSort')
plt.scatter(besttime[:, 0],  besttime[:,5], c='m', marker='<', alpha=0.9, label='HeapSort')
plt.scatter(besttime[:, 0],  besttime[:,6], c='y', marker='v', alpha=0.9, label='QuickSort')
plt.scatter(besttime[:, 0],  besttime[:,7], c='c', marker='>', alpha=0.9, label='MergeSort')
plt.legend()
plt.title('Computation time of some sorting algorithms - BESTCASE')
plt.xlabel('number of items (n)')
plt.ylabel('msec')
plt.savefig('bestcase.png')
plt.show()
```
